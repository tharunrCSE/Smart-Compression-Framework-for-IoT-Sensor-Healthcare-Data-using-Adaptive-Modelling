```python
import pandas as pd
import time
import os
import matplotlib.pyplot as plt
import pickle
```

```python
def lzw_compress(uncompressed: str, print_steps=False):
    """Compress a string to a list of output symbols using LZW."""
    dict_size = 256
    dictionary = {chr(i): i for i in range(dict_size)}
    w = ""
    result = []

    # Store dictionary growth for printing
    dictionary_log = {}

    for c in uncompressed:
        wc = w + c
        if wc in dictionary:
            w = wc
        else:
            result.append(dictionary[w])
            dictionary[wc] = dict_size
            dictionary_log[wc] = dict_size  # New entry added

            # Print each dictionary entry (optional)
            if print_steps:
                print(f"Added to dictionary: {repr(wc)} -> {dict_size}")

            dict_size += 1
            w = c

    if w:
        result.append(dictionary[w])

    return result, dictionary_log
```

```python
def lzw_decompress(compressed: list) -> str:
    """Decompress a list of output symbols to a string."""
    dict_size = 256
    dictionary = {i: chr(i) for i in range(dict_size)}
    result = []

    w = chr(compressed.pop(0))
    result.append(w)

    for k in compressed:
        if k in dictionary:
            entry = dictionary[k]
        elif k == dict_size:
            entry = w + w[0]
        else:
            raise ValueError(f"Bad compressed k: {k}")

        result.append(entry)
        dictionary[dict_size] = w + entry[0]
        dict_size += 1
        w = entry

    return "".join(result)
```

```python
input_file = "human_vital_signs_dataset_2024.csv"
df = pd.read_csv(input_file)
csv_data = df.to_csv(index=False)

compressed_file = "compressed_data.lzw"
decompressed_file = "decompressed_data.csv"

# Original size
original_size = len(csv_data.encode("utf-8"))

# -------- Compression --------
start_time = time.time()
compressed_data, dictionary_log = lzw_compress(csv_data, print_steps=False)
compression_time = time.time() - start_time

with open(compressed_file, "wb") as f:
    pickle.dump(compressed_data, f)

compressed_size = os.path.getsize(compressed_file)

# -------- Decompression --------
start_time = time.time()
with open(compressed_file, "rb") as f:
    loaded_data = pickle.load(f)
decompressed_data = lzw_decompress(loaded_data.copy())
decompression_time = time.time() - start_time

with open(decompressed_file, "w", encoding="utf-8") as f:
    f.write(decompressed_data)

decompressed_size = os.path.getsize(decompressed_file)

compression_ratio = original_size / compressed_size if compressed_size != 0 else 0

# ==========================
#   VERIFY
# ==========================
if csv_data.strip() == decompressed_data.strip():
    print("Data Verified: Original and Decompressed contents match.")
else:
    print("Data mismatch detected!")


# ==========================
#   PRINT DICTIONARY DETAILS
# ==========================
print("\n--- LZW DICTIONARY STATISTICS ---")
print(f"Initial dictionary size: 256")
print(f"New entries learned during compression: {len(dictionary_log)}")
print(f"Final dictionary size: {256 + len(dictionary_log)}")

print("\nDictionary Entries:")
for key, val in list(dictionary_log.items())[:500]:
    print(f"{val}: {repr(key)}")

if len(dictionary_log) > 500:
    print(f"... {len(dictionary_log) - 500} more entries not shown ...")


# ==========================
#   PRINT RESULTS
# ==========================
print("\n--- RESULTS (Using LZW Algorithm) ---")
print(f"Original File: {input_file}")
print(f"Compressed File: {compressed_file}")
print(f"Decompressed File: {decompressed_file}")
print(f"Original File Size: {original_size} bytes")
print(f"Compressed File Size: {compressed_size} bytes")
print(f"Decompressed File Size: {decompressed_size} bytes")
print(f"Compression Ratio: {compression_ratio:.2f}")
print(f"Compression Time: {compression_time:.4f} seconds")
print(f"Decompression Time: {decompression_time:.4f} seconds")

# ==========================
#   VISUALIZATION
# ==========================
plt.figure(figsize=(10, 4))

# File Size Comparison
plt.subplot(1, 2, 1)
plt.bar(['Original', 'Compressed'], [original_size, compressed_size], color=['orange', 'purple'])
plt.title('File Size Comparison')
plt.ylabel('Bytes')

# Time Comparison
plt.subplot(1, 2, 2)
plt.bar(['Compression', 'Decompression'], [compression_time, decompression_time], color=['blue', 'green'])
plt.title('Time Comparison')
plt.ylabel('Seconds')

plt.suptitle('LZW Algorithm Performance')
plt.tight_layout()
plt.show()
```
